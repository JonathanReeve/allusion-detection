```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('log.txt')
```

```python
def getDate(filename): 
    splitFile = filename.split('_')
    return splitFile[1]

def getAuthor(filename):
    splitFile = filename.split('_')
    return splitFile[2], splitFile[3]

def getTitle(filename):
    splitFile = filename.split('_')
    return splitFile[4]
```

```python
df['Date'] = df['TextB'].apply(getDate)
df['Author'] = df['TextB'].apply(getAuthor)
df['Title'] = df['TextB'].apply(getTitle)
df = df.sort_values(by='Date')
df = df.set_index('Date')
```

```python
sdf = df[['Author', 'Title', 'Nationality', 'Gender', 'Score']]
sdf
```

```python
def translateColor(color): 
    if color == 'M': 
        return 'b'
    else: 
        return 'r'
```

```python
sdf['Colors'] = sdf['Gender'].apply(translateColor)
```

```python
sdf
```

```python
mycolors = list(sdf['Colors'])
mycolors
```

```python
sdf.plot(figsize=(16,6), color=mycolors, kind='bar')
```
