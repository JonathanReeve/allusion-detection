```python
import pandas as pd
%matplotlib inline
```

```python
books = pd.read_csv('kjv-books-log.txt')
```

```python
books['kjv-book'] = books['kjv-book'].apply(lambda x: x.split('/')[2])
books['kjv-book'] = books['kjv-book'].apply(lambda x: x.split('.')[0])
```

```python
books
```

```python
bybooks = books.groupby('kjv-book').sum()
bybooks = bybooks['matches'] # remove unnecessary columns
bybooks = bybooks.sort_values(ascending=False)
```
