This notebook computes the most-referenced classical names in the CENLab corpus. 

```python
import re
```

```python
clas = open('classical-experiment.txt').read()
```

```python
clas[:100]
```

```python
clines = clas.split('\n')
```

```python
t = clines[0]
t
```

```python
def getTextWords(line): 
    text = re.findall(r'/texts/(.*?.txt)', line)
    words = re.findall('\x1b\[01;31m\x1b\[K(.*?)\x1b\[m\x1b\[K', line)
    return text, words
```

```python
clean = [getTextWords(line) for line in clines]
```

```python
figureDict = {}
for line in clean: 
    if len(line[1]) == 1: 
        if line[1][0] not in figureDict: 
            figureDict[line[1][0]] = 1
        else: # already there. Increment
            figureDict[line[1][0]] += 1
    else: 
        for figure in line[1]: 
            if figure not in figureDict: 
                figureDict[figure] = 1
            else: 
                figureDict[figure] += 1
```

```python
figureDict
```

```python
for figure in sorted(figureDict, key=figureDict.get, reverse=True):
  print(figure, figureDict[figure])
```
