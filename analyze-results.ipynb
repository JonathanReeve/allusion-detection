```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('txtlab-log.txt')
```

```python
def getDate(filename): 
    splitFile = filename.split('_')
    return int(splitFile[1])

def getAuthor(filename):
    splitFile = filename.split('_')
    return splitFile[2], splitFile[3]

def getTitle(filename):
    splitFile = filename.split('_')
    return splitFile[4]
```

```python
df['Date'] = df['TextB'].apply(getDate)
df['Author'] = df['TextB'].apply(getAuthor)
df['Title'] = df['TextB'].apply(getTitle)
df = df.sort_values(by='Date')
df = df.set_index('Date')
```

```python
sdf = df[['Author', 'Title', 'Nationality', 'Gender', 'Score']]
sdf
```

```python
def translateColor(color): 
    if color == 'B': 
        return 'b'
    else: 
        return 'r'
```

```python
sdf
```

```python
sdf.plot(sdf.index, 'Score', figsize=(16,6), color=sdf['Nationality'].apply(translateColor), kind='bar')
```

```python
g = sdf.groupby((sdf.index//10)*10).sum()
```

```python
g.plot(figsize=(14,6))
```

```python
def plotGroup(df): 
    g = df.groupby((df.index//10)*10).sum()
    g.plot(figsize=(14,6))
```

```python
brits = sdf.loc[sdf['Nationality'] == 'B']
americans = sdf.loc[sdf['Nationality'] == 'A']
men = sdf.loc[sdf['Gender'] == 'M']
women = sdf.loc[sdf['Gender'] == 'F']
```

```python
plotGroup(brits)
```

```python
plotGroup(americans)
```

```python
plotGroup(men)
```
